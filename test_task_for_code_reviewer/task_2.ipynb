```python
import pandas as pd
```

```python
#data
event_data = pd.read_csv('event_data.csv')
display(event_data.head())
```

#### 2. Highlight user cohorts based on the week of registration in the application. 

The cohort identifier should be the week ordinal (for example, the week from July 29 to August 4
should have identifier 31).

```python
event_data['event_date']= pd.to_datetime(event_data['event_date'])
event_data.info()
```

```python
event_data['cohort_week_id'] = event_data['event_date'].dt.week
display(event_data.head())
```

***Now we have information about event week***

#### 3. How many unique users in the cohort with ID 33?

```python
print('Unique users in the cohort with ID 33:', len(event_data.query('cohort_week_id==33').user_id.unique()))
```

#### 4. For each event, highlight the indicator lifetime - the weekly lifetime of the cohort. 

The lifetime indicator is calculated based on the serial number of the week in which the event
is committed, relative to the week of registration. For example, an event committed on
August 3 by a user from a cohort of registrants at 31 weeks will be committed on the zero
week of lifetime, and an event committed by the same user on August 5 will be committed
on the first week of lifetime).

```python
#information about first event date
first_event_date_for_users = event_data.groupby('user_id')['event_date'].min() 
first_event_date_for_users.name = 'first_event_date'
display(first_event_date_for_users.head()) 
```

***create and rename the information about first event date***

```python
#table with information about first event date
event_data = event_data.join(first_event_date_for_users,on='user_id')
display(event_data.head()) 
```

***create a table with new information***

```python
#check data types
event_data.info()
```

```python
#create column with first week of event
event_data['first_event_week'] = event_data['first_event_date'].dt.week
display(event_data.head()) 
```

***Now we have information about first week of event***

#### 5. Build a summary table of changes in the Retention Rate for cohorts depending on lifetime.

```python
#create column with lifetime
event_data['lifetime'] = event_data['cohort_week_id']-event_data['first_event_week']
display(event_data.head()) 
```

***Calculatred lifetime***

```python
#table with cohorts and unique users
cohorts = event_data.groupby(['first_event_week','lifetime']).agg({'user_id':'nunique'}).reset_index() 
display(cohorts)
```

***create table with any week for lifetime and unique number of users***

```python
#count users in cohorts
users_count = cohorts[cohorts['lifetime'] == 0][['first_event_week','user_id']]
users_count = users_count.rename(columns={'user_id':'cohort_users'}) 
print(users_count) 
```

```python
#create column with retention
cohorts = cohorts.merge(users_count,on='first_event_week') 
cohorts['retention'] = cohorts['user_id']/cohorts['cohort_users'] 
display(cohorts)
```

***create table with retention***

```python
#final table for retention rate
retention_rate = cohorts.pivot_table(index='first_event_week',columns='lifetime',values='retention',aggfunc='sum').reset_index()
display(retention_rate)
```

***Now we have pivot table with first event week, lifetime week and retention.***

#### 6. What is the 3 week retention rate for a cohort with ID 32? 

Give the answer in percent,rounded to 2 decimal places, inclusive.

```python
print('The 3 week retention rate for a cohort with ID 32:', round(retention_rate.xs(1)[3]*100,2))
```

#### 7. Build a summary table of changes in the indicator ARPPU (Average Revenue Per Paying User) for cohorts depending on lifetime.

```python
#create table with mean purchase on unique user on first week of event
purchases = event_data.groupby(['first_event_week','lifetime']).agg({'user_id':'nunique', 'purchase_amount': 'mean'}).reset_index() 
display(purchases)
```

***Create table with mean purchase amount, first event week, lifetime week and number of unique users***

```python
#final table for average revenue per paying user
average_revenue = purchases.pivot_table(index='first_event_week',columns='lifetime',values='purchase_amount',aggfunc='mean').reset_index()
display(average_revenue)
```

***Now we have pivot table with first event week, lifetime week and average revenue per paying user***

#### 8. What is the 3-week ARPPU of a cohort with ID 31? 

Give the answer with a floating point number, rounded to 2 decimal places, inclusive.

```python
print('The 3-week ARPPU of a cohort with ID 31:', round(average_revenue.xs(0)[3],2))
```

#### 9. What is the median time between user registration and first purchase? 

Give the answer in seconds (!) As an integer.

```python
#create a column with time in seconds
event_data['time_between_purchase']=(event_data['event_date']-event_data['first_event_date']).dt.seconds
display(event_data.head())
```

```python
print('Median time between user registration and first purchase:', round(event_data.time_between_purchase.median()), 'seconds')
```
