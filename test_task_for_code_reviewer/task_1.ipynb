#### 1.5. Create a movie_duration_category column, 
which will contain three categories depending on the film length:

    â€¢ Category "1. <90" if the film is less than 90 minutes long
    â€¢ Category "2. 90â€“120" if the film is between 90 minutes and two hours long (inclusively)
    â€¢ Category "3. >120" if the film is more than two hours long

```python
def movie_category_func(row):
    """Function to determine the movie category.

    Input - a string with the length of the movie. 
    Output - movie category,
    """  
    
    duration = row['duration'] #a string with the length of the movie
   
    if duration <90: # if duration is less than 90 minutes long
            return 1 #output 1
    if duration >120: #if duration is more than two hours long
            return 3 #output 3
    return 2 #if duration is between 90 minutes and two hours long, output 2

#create column with category
movie_metadata['movie_duration_category'] = movie_metadata.apply(movie_category_func, axis=1)

display(movie_metadata.head())
```

***Now we have column with information about movie duration category***

#### 1.6. Build a summary table for films released after 2000 (inclusively), to list the numbers of films:

    â€¢ Table rows: year
    â€¢ Table columns: movie duration category ("<90", "90â€“120", ">120")
    â€¢ The year of release should be displayed in the YYYY format.

```python
#create table with year after 2000, categories and number of movies
count_movie_category = (movie_metadata.query('title_year >= 2000')
                        .pivot_table(index='title_year', columns='movie_duration_category',
                        aggfunc={'movie_duration_category':'count'})).reset_index()
count_movie_category.columns = ['title_year', 'category 1', 'category 2' , 'category 3']
display(count_movie_category)
```

***Now we have table with information about number of movies in categories***

#### 1.7. How many films between 90 minutes and two hours long were released in 2008?

```python
print(count_movie_category.xs(8)[2], 'films between 90 minutes and two hours long were released in 2008') 
```

#### 1.8. The plot_keywords column holds keywords characterizing the film's plot. 

Using the data in this column, create a column called movie_plot_category, to contain four categories
depending on the key words in the column:
    
    â€¢ Category "love_and_death" if the keywords include both "love" and "death"
    â€¢ Category "love" if the keywords include the word "love"
    â€¢ Category "death" if the keywords include the word "death"
    â€¢ Category "other" if the keywords do not meet the conditions above

```python
#we have nulls, so we need to replace them to undifirend
movie_metadata['plot_keywords'] = movie_metadata['plot_keywords'].fillna('undifirend')
```

```python
def plot_keywords_func(row):
    """Function to determine the plot category.

    Input - a string with the plot keywords. 
    Output - plot category,
    """  
    plot_keywords = row['plot_keywords'] #a string with the plot keywords
   
    if 'love' in plot_keywords and 'death' in plot_keywords: #if the keywords include both "love" and "death"
            return 'love_and_death' #output love_and_death
    if 'love' in plot_keywords: #if the keywords include the word "love"
            return 'love' #output love
    if 'death' in plot_keywords: #if the keywords include the word "death"
        return 'death' #output death
    return 'other' #if the keywords do not meet the conditions above, output other

#create column with plot category
movie_metadata['movie_plot_category'] = movie_metadata.apply(plot_keywords_func, axis=1)

display(movie_metadata.head())
```

***Now we have column with information about movie plot category***

#### 1.9. The imdb_score column shows a viewer rating for the film. 

Build a table to reflect the average rating of films depending on which movie_plot_category category they belong to.

```python
#create table with average rating of films in plot category
count_rating_category = (movie_metadata.pivot_table(index='movie_plot_category', values='imdb_score',
                        aggfunc={'imdb_score':'mean'})).reset_index()
display(count_rating_category)
```

***Now we have table with information about imdb score in movie plot categories***

#### 1.10. What is the average rating of films in the "love" category? 

Give the answer as a floatingpoint figure rounded to two decimal places.

```python
print('The average rating of films in the "love" category:', round(count_rating_category.xs(1)[1],2))
```

What is the median budget for all the films listed? Give the answer as an integer.
